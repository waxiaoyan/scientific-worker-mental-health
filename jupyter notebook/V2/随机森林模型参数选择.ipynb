```python
fpath = r"/Users/wangxiaoyan/Documents/wy/论文/git/scientific-worker-mental-health/data/V2/csv/2017-全国科技工作者-焦虑抑郁-维度数据.csv"
data = pd.read_csv(fpath)
random_state=42
y = data['depression']
x = data.drop(labels=['depression','anxiety'], axis=1)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=random_state)
# 过采样
smote = SMOTEENN(random_state=random_state)
x_train_resampled, y_train_resampled = smote.fit_resample(x_train, y_train)
#设置XGBClassifier的参数范围
params = {
    'max_depth': [2, 3, 4], 
    'min_child_weight': [1, 3, 5],
    'scale_pos_weight': [1, 5, 10],
    'n_estimators': [100, 200, 300], 
    'learning_rate': [0.1, 0.01, 0.05] 
}

# 设置GridSearchCV
grid_search = GridSearchCV(estimator = XGBClassifier(objective ='multi:softprob'), 
                           param_grid = params,  
                           scoring = 'f1_macro',  
                           cv = 5,
                           n_jobs = -1)

# 进行网格搜索           
grid_search.fit(x_train_resampled, y_train_resampled)

# 查看最佳参数和最佳分数
best_params = grid_search.best_params_  
best_score = grid_search.best_score_  
print("best params:")
print(best_params)
print(best_score)
```
