```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
# import xgboost as xgb
import matplotlib.pyplot as plt
from sklearn.model_selection import GridSearchCV
import pandas as pd, numpy as np
import matplotlib as mpl
# import lightgbm as lgb
from sklearn.metrics import mean_squared_error
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.ensemble import RandomForestClassifier
```

```python
# fpath ="/Users/wangxiaoyan/Documents/wy/论文/git/scientific-worker-mental-health/data/all/2017科技人员心理健康-清洗处理后数据.csv"
fpath = r"/Users/wangxiaoyan/Documents/wy/论文/git/scientific-worker-mental-health/data/all/2017科技人员心理健康-排除背景信息-空值填充处理后数据.csv"
Scientific_Sample=pd.read_csv(fpath)
#个人创新行为量表1
IIS_columns = Scientific_Sample.filter(like='IIS', axis=1)
Scientific_Sample['Sum_IIS'] = IIS_columns.sum(axis=1)

#匹兹堡睡眠质量指数2
PSQI_columns = Scientific_Sample.filter(like='PSQI', axis=1)
Scientific_Sample['Sum_PSQI'] = PSQI_columns.sum(axis=1)

#工作倦怠量表3
MBIGS_columns = Scientific_Sample.filter(like='MBIGS', axis=1)
Scientific_Sample['Sum_MBIGS'] = MBIGS_columns.sum(axis=1)
#工作家庭冲突量表4
WFCS_columns = Scientific_Sample.filter(like='WFCS', axis=1)
Scientific_Sample['Sum_WFCS'] = WFCS_columns.sum(axis=1)

#幸福感与工作满意度5
LSS_columns = Scientific_Sample.filter(like='LSS', axis=1)
Scientific_Sample['Sum_LSS'] = LSS_columns.sum(axis=1)


#科技工作者问卷（自编）10
STPQ_columns = Scientific_Sample.filter(like='STPQ', axis=1)
Scientific_Sample['Sum_STPQ'] = STPQ_columns.sum(axis=1)


#科技工作者问卷-岗位（自编）12
STPQpost_columns = Scientific_Sample.filter(like='STPQpost', axis=1)
Scientific_Sample['Sum_STPQpost'] = STPQpost_columns.sum(axis=1)

#科技工作者问卷-工作时间的分配（自编）13
STPQawt_columns = Scientific_Sample.filter(like='STPQawt', axis=1)
Scientific_Sample['Sum_STPQawt'] = STPQawt_columns.sum(axis=1)

#科技工作者问卷-工作满意度（自编）14
STPQjs_columns = Scientific_Sample.filter(like='STPQjs', axis=1)
Scientific_Sample['Sum_STPQjs'] = STPQjs_columns.sum(axis=1)

#科技工作者问卷-收入水平（自编）15
STPQil_columns = Scientific_Sample.filter(like='STPQil', axis=1)
Scientific_Sample['Sum_STPQil'] = STPQil_columns.sum(axis=1)

#科技工作者问卷-生活困扰（自编）16
STPQlt_columns = Scientific_Sample.filter(like='STPQlt', axis=1)
Scientific_Sample['Sum_STPQlt'] = STPQlt_columns.sum(axis=1)


#科技工作者问卷-职业评价（自编）20
STPQpe_columns = Scientific_Sample.filter(like='STPQpe', axis=1)
Scientific_Sample['Sum_STPQpe'] = STPQpe_columns.sum(axis=1)

#科技工作者问卷-职业评价（自编）22
OSI_columns = Scientific_Sample.filter(like='OSI', axis=1)
Scientific_Sample['Sum_OSI'] = OSI_columns.sum(axis=1)
###########################空值大于50%##########################
#科技工作者问卷-社会资本（自编）17
STPQsc_columns = Scientific_Sample.filter(like='STPQsc', axis=1)
Scientific_Sample['Sum_STPQsc'] = STPQsc_columns.sum(axis=1)

#科技工作者问卷-组织氛围（自编）18
STPQoc_columns = Scientific_Sample.filter(like='STPQoc', axis=1)
Scientific_Sample['Sum_STPQoc'] = STPQoc_columns.sum(axis=1)

# #科技工作者问卷-职业价值观（自编）19
STPQpv_columns = Scientific_Sample.filter(like='STPQpv', axis=1)
Scientific_Sample['Sum_STPQpv'] = STPQpv_columns.sum(axis=1)

#心理健康服务现状与需求6
CSDMHS_columns = Scientific_Sample.filter(like='CSDMHS', axis=1)
Scientific_Sample['Sum_CSDMHS'] = CSDMHS_columns.sum(axis=1)

#心理健康素养问卷7
MHLQ_columns = Scientific_Sample.filter(like='MHLQ', axis=1)
Scientific_Sample['Sum_MHLQ'] = MHLQ_columns.sum(axis=1)

#心理求助障碍量表8
PBSMHS_columns = Scientific_Sample.filter(like='PBSMHS', axis=1)
Scientific_Sample['Sum_PBSMHS'] = PBSMHS_columns.sum(axis=1)

#心理资本问卷9
PCQ_columns = Scientific_Sample.filter(like='PCQ', axis=1)
Scientific_Sample['Sum_PCQ'] = PCQ_columns.sum(axis=1)

#科技工作者问卷-行政干预（自编）21
STPQai_columns = Scientific_Sample.filter(like='STPQai', axis=1)
Scientific_Sample['Sum_STPQai'] = STPQai_columns.sum(axis=1)

#科技工作者问卷-创业态度（自编）11
STPQea_columns = Scientific_Sample.filter(like='STPQea', axis=1)
Scientific_Sample['Sum_STPQea'] = STPQea_columns.sum(axis=1)

```

```python
Scientific_Sample.drop(columns=Scientific_Sample.filter(regex='^IIS').columns, inplace=True)
Scientific_Sample.drop(columns=Scientific_Sample.filter(regex='^PSQI').columns, inplace=True)
Scientific_Sample.drop(columns=Scientific_Sample.filter(regex='^MBIGS').columns, inplace=True)
Scientific_Sample.drop(columns=Scientific_Sample.filter(regex='^STPQjs').columns, inplace=True)
Scientific_Sample.drop(columns=Scientific_Sample.filter(regex='^STPQil').columns, inplace=True)
Scientific_Sample.drop(columns=Scientific_Sample.filter(regex='^STPQlt').columns, inplace=True)
Scientific_Sample.drop(columns=Scientific_Sample.filter(regex='^STPQsc').columns, inplace=True)
Scientific_Sample.drop(columns=Scientific_Sample.filter(regex='^STPQoc').columns, inplace=True)
Scientific_Sample.drop(columns=Scientific_Sample.filter(regex='^STPQpv').columns, inplace=True)
Scientific_Sample.drop(columns=Scientific_Sample.filter(regex='^STPQpe').columns, inplace=True)
Scientific_Sample.drop(columns=Scientific_Sample.filter(regex='^STPQai').columns, inplace=True)
Scientific_Sample.drop(columns=Scientific_Sample.filter(regex='^OSI').columns, inplace=True)
Scientific_Sample.drop(columns=Scientific_Sample.filter(regex='^WFCS').columns, inplace=True)
Scientific_Sample.drop(columns=Scientific_Sample.filter(regex='^LSS').columns, inplace=True)
Scientific_Sample.drop(columns=Scientific_Sample.filter(regex='^CSDMHS').columns, inplace=True)
Scientific_Sample.drop(columns=Scientific_Sample.filter(regex='^MHLQ').columns, inplace=True)
Scientific_Sample.drop(columns=Scientific_Sample.filter(regex='^PBSMHS').columns, inplace=True)
Scientific_Sample.drop(columns=Scientific_Sample.filter(regex='^PCQ\d').columns, inplace=True)
Scientific_Sample.drop(columns=Scientific_Sample.filter(regex='^STPQ').columns, inplace=True)
Scientific_Sample.drop(columns=Scientific_Sample.filter(regex='^STPQea').columns, inplace=True)
Scientific_Sample.drop(columns=Scientific_Sample.filter(regex='^STPQpost').columns, inplace=True)
Scientific_Sample.drop(columns=Scientific_Sample.filter(regex='^STPQawt').columns, inplace=True)
```
