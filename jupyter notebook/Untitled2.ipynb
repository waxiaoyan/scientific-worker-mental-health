```python
import pandas as pd, numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("/Users/wangxiaoyan/Documents/wy/论文/git/scientific-worker-mental-health/data/科技人员焦虑表-重要特征数据.csv")
corr_matrix = data.corr()

corr_with_target = corr_matrix['anxiety']
corr_with_target = corr_with_target[np.abs(corr_with_target) > 0.6]
plt.figure(figsize=(6, 4))
sns.heatmap(data[corr_with_target.index].corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix (with Target)')
plt.show()
```
