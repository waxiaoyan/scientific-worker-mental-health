```python
import pandas as pd
```

```python
fpath = r"/Users/wangxiaoyan/Documents/wy/论文/git/scientific-worker-mental-health/data/all/2017-科技人员心理健康SMOTEENN特征筛选后-交集特征-数据集.csv"

df = pd.read_csv(fpath)
to_remove = pd.DataFrame()

# 遍历每个类别
for category in [0,1,2,3]:
    # 获取类别为 category 的数据行
    category_rows = df[df['anxiety'] == category]
    
    # 获取 category 类别下的30%的数据行
    num_rows = int(0.3 * len(category_rows))
    rows_to_move = category_rows.sample(n=num_rows, random_state=42)
    
    # 将 rows_to_move 添加到 to_remove 中
    to_remove = pd.concat([to_remove, rows_to_move])
    
    # 从原始数据中删除 rows_to_move
    df = df.drop(rows_to_move.index)
    
# 将 to_remove 存储到文件中
df.to_csv('/Users/wangxiaoyan/Documents/wy/论文/git/scientific-worker-mental-health/data/all/2017-科技人员心理健康SMOTEENN特征筛选后-交集焦虑类别70.csv', index=False)
to_remove.to_csv('/Users/wangxiaoyan/Documents/wy/论文/git/scientific-worker-mental-health/data/all/2017-科技人员心理健康SMOTEENN特征筛选后-交集焦虑类别30.csv', index=False)

```
